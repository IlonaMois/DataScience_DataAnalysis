### Расчёт эффективности

Необходимо смоделировать процесс восстановления золота из золотосодержащей руды.

Эффективность обогащения рассчитывается по формуле:

$$Recovery = \frac{C*\left ( F - T \right )}{F*\left ( C - T \right )}*100\%$$

где:
- C — доля золота в концентрате после флотации/очистки;
- F — доля золота в сырье/концентрате до флотации/очистки;
- T — доля золота в отвальных хвостах после флотации/очистки.

Для прогноза коэффициента необходимо найти долю золота в концентратах и хвостах. Причём важен не только финальный продукт, но и черновой концентрат. 
Правильность расчета эффективности восстановления, определим метрикой MAE (среднюю абсолютную ошибку) между нашими расчётами и значением признака `rougher.output.recovery`

```python
C = gold_full['rougher.output.concentrate_au']
F = gold_full['rougher.input.feed_au'] 
T = gold_full['rougher.output.tail_au']

gold_full['rougher.output.recovery_calc'] = ((C*(F-T))/(F*(C-T)))*100
mae = mean_absolute_error(gold_full['rougher.output.recovery'],gold_full['rougher.output.recovery_calc'])
gold_full.drop('rougher.output.recovery_calc', axis=1,inplace=True)

print('Средняя абсолютная ошибка MAE эффективности восстановления равна:', mae)
```

**Вывод:**
 
 Значение Средней абсолютной ошибки MAE чрезвычайно мало. Исходные расчеты эффективности восстановления верны.

## Предобработка данных 

**Устранение пропусков и установка индексов**

```python
# В условиях задачи сказано, что замеры параметров производятся по времени.
# Близкие по времени измерения признаки похожи между собой. Заполним пропуски в данных значениями, ближайшими к пропуску

def preprocessing(data):
    data.fillna(method = 'ffill', axis = 0, inplace = True)
    data.set_index('date',inplace = True)

preprocessing(gold_train)
preprocessing(gold_test)
preprocessing(gold_full)
```

```python
gold_test.head()
```

**Восстановление столбцов**
Чтобы избежать мультиколлинеарности данных и одновременно не удалить важные признаки, необходимо воспользоваться методом VIF. 

```python
# Рассчитаем VIF для колонок исходного дата-сета:

x_temp = sm.add_constant(gold_full)

vif = pd.DataFrame()
vif["vif_factor"] = [variance_inflation_factor(x_temp.values, i) for i in range (x_temp.values.shape[1])]
vif["features"] = x_temp.columns

# Найдем колонки, в которых значение VIF будет больше 8:
vif_hight = vif.query('vif_factor >=8')
vif_hight
```

```python
# Создаем список столбцов, подлежащих удалению и избавляемся от константы:
delate_columns_list = vif_hight['features']
delate_columns_list = delate_columns_list[delate_columns_list != 'const']
delate_columns_list
```

```python
# Сделаем наши обучающую и тестовую выборку одинаковыми, исключив колонки с высоким коэффициентом VIF

gold_test = gold_test.reset_index().merge(gold_full, how = 'inner').set_index('date')

gold_test = gold_test.drop(columns=delate_columns_list,axis=1)
gold_train = gold_train.drop(columns=delate_columns_list,axis=1)
```

```python
print('Количество столбцов в Обучающей выборке после удаления мультиколлинеарных признаков: ',gold_train.shape)
print('Количество столбцов в Тестовой выборке после удаления мультиколлинеарных признаков: ',gold_test.shape)
```

```python
gold_test.head()
```

**Вывод предобработки данных**
- В ходе предобработки, данные были проиндексированы, а так же устранеы пропуски методом ffill
- В качестве борьбы с мультиколлинеарностью был выбран метод VIF. Используя данный метод, были получены признаки, с наибольшим коэффициентом VIF (признаки, подлежащие удалению). В результате, Обучающая и Тестовая выборки имеют 42 признака, включая целевые признаки.

### Построение и обучение модели

#### sMAPE

*Метрика качества, необходимая для оценки эффективности процесса очистки золотосодержащей руды — sMAPE (симметричное среднее абсолютное процентное отклонение).  
Формула вычисления метрики:*


$$sMAPE = \frac{1}{N} \sum_{i=1}^{N}\frac{\left | y_{i} -\
\hat{y}_{i}\right |}{\left ( \left | y_{i} \right | +\left | \hat{y}_{i} \right |\right )/2} *100\%$$  


- *Значение целевого признака для объекта с порядковым номером i в выборке, на которой измеряется качество:* $$y_{i}$$   
- *Значение предсказания для объекта с порядковым номером i, например, в тестовой выборке:* $$\hat{y}_{i}$$  
- *Количество объектов в выборке:* $$N$$  
- *Суммирование по всем объектам выборки (i меняется от 1 до N).* $$\sum_{i=1}^{N}$$


**Формула итогового sMAPE**
$$Итоговое sMAPE = 25\%*sMAPE(rougher)+75\%*sMAPE(final)$$

```python
# Построим функцию, которая возвращает sMAPE:
def smape(target, prediction): 
    diff = abs(target - prediction)
    add = (abs(target) + abs(prediction))/2
    smape_stage = 1/len(target)*sum(diff/add)*100
    
    return smape_stage

# Построим функцию, которая возвращает итоговую sMAPE:
def final_smape(smape_r,smape_f):
    final_smape = 0.25*smape_r + 0.75*smape_f
    return final_smape
```

```python
# В каждой из выборок (обучающая и тестовая) выделим признаки и целевой признак
features_train = train_final.drop(['rougher.output.recovery','final.output.recovery'], axis=1)
target_rougher_train = train_final['rougher.output.recovery']
target_final_train = train_final['final.output.recovery']


    
features_test = test_final.drop(['rougher.output.recovery','final.output.recovery'], axis=1)
target_rougher_test = test_final['rougher.output.recovery']
target_final_test = test_final['final.output.recovery']
```

```python
# Найдем оптимальные гиперпараметры для моделей дерево решений и случайного леса:
linear_model = LinearRegression()

tree_params = {'max_depth': range(1,20,1)}
tree_model = DecisionTreeRegressor(random_state=12345)

forest_params = {'max_depth': range(5,20,1),'n_estimators': range(135,139,1)}
forest_model = RandomForestRegressor(random_state=12345)
```

```python
# Создадим оценку которая будет отображать SMAP:
score_smape = make_scorer(smape, greater_is_better=False)
```

**Linear Regression:**

```python
%%time
cv_sc_lin_rougher = cross_val_score(LinearRegression(),
                                          features_train,
                                          target_rougher_train,
                                          cv=5, scoring=score_smape).mean()
cv_sc_lin_final = cross_val_score(LinearRegression(),
                                        features_train,
                                        target_final_train,
                                        cv=5, scoring=score_smape).mean()
smape_lin = final_smape(cv_sc_lin_rougher,cv_sc_lin_final)
print(f"Итоговое симметричное среднее абсолютное процентное отклонение: {smape_lin:.3f}%")
```

**Decision Tree Regressor**

```python
%%time

tree_grid_rougher = GridSearchCV(estimator=tree_model, param_grid=tree_params, cv=5, scoring=score_smape)
tree_grid_rougher.fit(features_train, target_rougher_train)
tree_grid_rougher_best = tree_grid_rougher.best_score_

tree_grid_final = GridSearchCV(estimator=tree_model, param_grid=tree_params, cv=5, scoring=score_smape)
tree_grid_final.fit(features_train, target_final_train)
tree_grid_final_best = tree_grid_final.best_score_

smape_tree = final_smape(tree_grid_rougher_best,tree_grid_final_best)
print(f"Итоговое симметричное среднее абсолютное процентное отклонение: {smape_tree:.3f}%")
```

**Random Forest Regressor**

```python
%%time

forest_grid_rougher = GridSearchCV(estimator=forest_model, param_grid=forest_params, cv=5, scoring=score_smape)
forest_grid_rougher.fit(features_train, target_rougher_train)
forest_grid_rougher_best = forest_grid_rougher.best_score_

forest_grid_final = GridSearchCV(estimator=forest_model, param_grid=forest_params, cv=5, scoring=score_smape)
forest_grid_final.fit(features_train, target_final_train)
forest_grid_final_best = forest_grid_final.best_score_

smape_forest = final_smape(forest_grid_rougher_best,forest_grid_final_best)
print(f"Итоговое симметричное среднее абсолютное процентное отклонение: {smape_forest:.3f}%")
```

**Вывод**

Наилучший результат Итоговой SMAPE демонстрирует модель **Random Forest Regressor** подобранными с помощью метода оптимизации подбора гиперпараметров GridSearchCV. **Итоговое SMPE: -6.902%**

### Проверка на адекватность

*Для проверки выбранной модели на адекватность, выбрана модель `DummyRegressor` библиотеки `Sklearn`.  
Этот классификатор служит простой базой для сравнения с другими более сложными классификаторами.  
Конкретное поведение базовой линии выбирается с помощью параметра `strategy`.*  
- *`mean`, всегда предсказывает среднее значение тренировочного набора*

```python
%%time
dummy_uniform = DummyRegressor(strategy='mean').fit(features_train, target_rougher_train)
dummy_rog = dummy_uniform.predict(features_test)
smape_dummy_roug = smape(target_rougher_test,dummy_rog)

dummy_uniform_1 = DummyRegressor(strategy='mean').fit(features_train, target_final_train)
dummy_fin = dummy_uniform_1.predict(features_test)
smape_dummy_fin = smape(target_final_test,dummy_fin)

smape_dummy = final_smape(smape_dummy_roug,smape_dummy_fin)

print(f"Итоговое симметричное среднее абсолютное процентное отклонение: {smape_dummy:.3f}%")
```

*Итог проверки на адекватность*
- Итоговое SMPE истинной модели Случайного Леса: 7.516%
- Итоговое SMPE случайной модели: 7.779%
*Истинная моедль Случайного Леса является более эффективной*
