```python
# Импортируем библиотеки и методы, необходымые для работы:
import pandas as pd
from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression 
from sklearn.dummy import DummyClassifier 

from sklearn import tree
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

Данные содержат количество совершенных звонков, их длительность в минутах, количество отправленнных смс сообщений, количество затраченных Мб и тип тарифа Ultra (1- Ultra, 0 - Smart). Качественным (целевым) признаком, будет являться колонка is_ultra*

```python
 # Разделим данные на сет признаков и целевой признак:
features = df.drop(['is_ultra'], axis=1)
target = df['is_ultra']
```

```python
# Разделим данные на обучающую, валидационную и тестовую выборки в пропорциях 50:25:25 :
features_train, features_test, target_train, target_test = train_test_split(
    features, target, stratify=target, train_size=0.5,random_state=12345)  # разделение на тренировочный сет и тестовый

features_test, features_valid, target_test, target_valid = train_test_split(
    features_test, target_test, stratify=target_train, train_size=0.5,random_state=12345) # разделение тестовый сет на обучающий и валидацинный
```

```python
# Проверим, сколько получилось данных в каждой выборке:
print('Размер обучающей выборки:',features_train.shape, target_train.shape)
print('Стратификация обучающей выборки:'),print(target_train.value_counts()/len(target_train)*100)
print(' ')
print('Размер валидационной выборки:', features_valid.shape, target_valid.shape)
print('Стратификация обучающей выборки:'),print(target_train.value_counts()/len(target_train)*100)
print(' ')
print('Размер тестовой выборки:', features_test.shape, target_test.shape)
print('Стратификация обучающей выборки:'),print(target_train.value_counts()/len(target_train)*100)
print(' ')
```

Разделение данных на обучающую, валидационную и тестовую выборки готово. Приступаем к тестированию качества разных моделей ML

**Дерево решений**

```python
%%time
# Определяем наилучшую глубину дерева решений с помощью функции, при которой модель будет выдавать максимально точные результаты:

best_model = None
best_result_tree = 0
best_depth = 0

for depth in range(1, 6):
    model_tree = DecisionTreeClassifier(random_state=12345, max_depth=depth) 
    model_tree.fit(features_train, target_train)        
    predicts = model_tree.predict(features_valid)
    result = accuracy_score(target_valid, predicts) 
    
    clf = tree.DecisionTreeClassifier(max_depth=depth)  # set hyperparameter
    clf.fit(features_train, target_train)

    # plot tree
    plt.figure(figsize=(12,12))  # set plot size (denoted in inches)
    tree.plot_tree(clf, fontsize=10)
    plt.show()
    
    if result > best_result_tree:
        best_model_tree = model_tree
        best_result_tree = result
        best_depth_tree = depth
        
print("Accuracy лучшей модели:", best_result_tree)
print('Глубина дерева:', best_depth_tree)
```

**Случайный Лес**

```python
%%time
best_model_forest = None
best_result_forest = 0
best_est_forest = 0
best_depth_forest = 0

for est in range(10,50,5):
     for depth in range (1, 11):
        model_forest = RandomForestClassifier(random_state=12345, n_estimators=est, max_depth=depth) 
        model_forest.fit(features_train, target_train) 
        predicts = model_forest.predict(features_valid)
        result = accuracy_score(target_valid, predicts) 
        if result > best_result_forest:
            best_model_forest = model_forest
            best_result_forest = result
            best_est_forest = est
            best_depth_forest = depth
        
print("Accuracy лучшей модели:", best_result_forest)
print('Количество деревьев:', best_est_forest)
print('Глубина дерева:', best_depth_forest)
```

**Логистическая регрессия**

```python
%%time
model_logistic = LogisticRegression(random_state=12345, solver='lbfgs', max_iter= 1000) 
model_logistic.fit(features, target)
 
result_logistic = model_logistic.score(features_valid,target_valid)
print("Accuracy модели логистческой регрессии:", result_logistic)
```

### Вывод:

Точность предсказания результатов моделей составляют:
 - Лес решений = 0.80

Скорость предсказания результатов:

Модель **Случайный Лес** наиболее точная, однако вместе с тем, на решение потребовалось в разы больше времени, чем у других моделей
