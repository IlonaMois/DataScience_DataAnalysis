## The process of doing work

```python
# Import libraries and methods necessary for work:
import pandas as pd
from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression 
from sklearn.dummy import DummyClassifier 

from sklearn import tree
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
# Import dataset:
try:
    df = pd.read_csv('/datasets/users_behavior.csv')
except:
    df = pd.read_csv('C:/Users/Админ/Desktop/Jupyter/Yandex/users_behavior.csv')
```

*According to the task, data preprocessing is not required. The data contains the number of calls made, their duration in minutes, the number of sent SMS messages, the amount of MB spent and the Ultra tariff type (1-Ultra, 0-Smart). A qualitative (target) feature will be the is_ultra* column

```python
# Divide the data into a set of features and a target feature:
features = df.drop(['is_ultra'], axis=1)
target = df['is_ultra']
```

```python
# Divide the data into training, validation and test sets in proportions 50:25:25 :
features_train, features_test, target_train, target_test = train_test_split(
    features, target, stratify=target, train_size=0.5,random_state=12345)  # разделение на тренировочный сет и тестовый

features_test, features_valid, target_test, target_valid = train_test_split(
    features_test, target_test, stratify=target_train, train_size=0.5,random_state=12345) # разделение тестовый сет на обучающий и валидацинный
```

```python
# Let's check how much data was obtained in each sample:
print('Training sample size:',features_train.shape, target_train.shape)
print('Stratification of the training sample:'),print(target_train.value_counts()/len(target_train)*100)
print(' ')
print('Validation sample size:', features_valid.shape, target_valid.shape)
print('Validation set stratification:'),print(target_train.value_counts()/len(target_train)*100)
print(' ')
print('Test sample size:', features_test.shape, target_test.shape)
print('Test set stratification:'),print(target_train.value_counts()/len(target_train)*100)
print(' ')
```

The division of data into training, validation and test sets is ready. Let's start testing the quality of different ML models

**Decision Tree**

```python
%%time
# Determine the best depth of the decision tree using the function at which the model will produce the most accurate results:
best_model = None
best_result_tree = 0
best_depth = 0

for depth in range(1, 6):
    model_tree = DecisionTreeClassifier(random_state=12345, max_depth=depth) 
    model_tree.fit(features_train, target_train)        
    predicts = model_tree.predict(features_valid)
    result = accuracy_score(target_valid, predicts) 
    
    clf = tree.DecisionTreeClassifier(max_depth=depth)  # set hyperparameter
    clf.fit(features_train, target_train)

    # plot tree
    plt.figure(figsize=(12,12))  # set plot size (denoted in inches)
    tree.plot_tree(clf, fontsize=10)
    plt.show()
    
    if result > best_result_tree:
        best_model_tree = model_tree
        best_result_tree = result
        best_depth_tree = depth
        
print("Accuracy of the best model:", best_result_tree)
print('Tree depth:', best_depth_tree)
```

**Random Forest**

```python
%%time
best_model_forest = None
best_result_forest = 0
best_est_forest = 0
best_depth_forest = 0

for est in range(10,50,5):
     for depth in range (1, 11):
        model_forest = RandomForestClassifier(random_state=12345, n_estimators=est, max_depth=depth) 
        model_forest.fit(features_train, target_train) 
        predicts = model_forest.predict(features_valid)
        result = accuracy_score(target_valid, predicts) 
        if result > best_result_forest:
            best_model_forest = model_forest
            best_result_forest = result
            best_est_forest = est
            best_depth_forest = depth
        
print("Accuracy of the best model:", best_result_forest)
print('Number of trees:', best_est_forest)
print('Tree depth:', best_depth_forest)
```

**Logistic regression**

```python
%%time
model_logistic = LogisticRegression(random_state=12345, solver='lbfgs', max_iter= 1000) 
model_logistic.fit(features, target)
 
result_logistic = model_logistic.score(features_valid,target_valid)
print("Accuracy of logistic regression model:", result_logistic)
```
