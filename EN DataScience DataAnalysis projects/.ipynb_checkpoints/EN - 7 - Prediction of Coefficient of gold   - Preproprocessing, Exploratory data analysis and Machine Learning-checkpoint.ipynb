### Efficiency calculation

It is necessary to simulate the process of recovering gold from gold ore.

Enrichment efficiency is calculated by the formula:

$$Recovery = \frac{C*\left ( F - T \right )}{F*\left ( C - T \right )}*100\%$$

where:
- C is the proportion of gold in the concentrate after flotation/cleaning;
- F is the proportion of gold in the raw material/concentrate before flotation/cleaning;
- T is the proportion of gold in tailings after flotation/cleaning.

To predict the coefficient, it is necessary to find the share of gold in concentrates and tailings. Moreover, not only the final product is important, but also the rough concentrate.
The correctness of the recovery efficiency calculation, we define the MAE metric (mean absolute error) between our calculations and the value of the feature `rougher.output.recovery`

```python
C = gold_full['rougher.output.concentrate_au']
F = gold_full['rougher.input.feed_au'] 
T = gold_full['rougher.output.tail_au']

gold_full['rougher.output.recovery_calc'] = ((C*(F-T))/(F*(C-T)))*100
mae = mean_absolute_error(gold_full['rougher.output.recovery'],gold_full['rougher.output.recovery_calc'])
gold_full.drop('rougher.output.recovery_calc', axis=1,inplace=True)

print('The mean absolute error MAE of recovery efficiency is:', mae)
```

**Conclusion:**
 
  The MAE value is extremely small. The initial recovery efficiency calculations are correct.

## Data preprocessing

**Remove missing values and set indexes**

```python
# In the conditions of the problem it is said that the parameters are measured in time.
# Signs close in time measurements are similar to each other. Fill missing values in data with values closest to the missing values
def preprocessing(data):
    data.fillna(method = 'ffill', axis = 0, inplace = True)
    data.set_index('date',inplace = True)

preprocessing(gold_train)
preprocessing(gold_test)
preprocessing(gold_full)
```

```python
gold_test.head()
```

**Restore columns**
To avoid data multicollinearity and at the same time not to remove important features, it is necessary to use the VIF method.

```python
# Calculate the VIF for the columns of the original dataset:

x_temp = sm.add_constant(gold_full)

vif = pd.DataFrame()
vif["vif_factor"] = [variance_inflation_factor(x_temp.values, i) for i in range (x_temp.values.shape[1])]
vif["features"] = x_temp.columns

# Find columns where the VIF value is greater than 8:
vif_hight = vif.query('vif_factor >=8')
vif_hight
```

```python
# Create a list of columns to be removed and remove constants:
delate_columns_list = vif_hight['features']
delate_columns_list = delate_columns_list[delate_columns_list != 'const']
delate_columns_list
```

```python
# Make our training and test sets the same by excluding columns with high VIF

gold_test = gold_test.reset_index().merge(gold_full, how = 'inner').set_index('date')

gold_test = gold_test.drop(columns=delate_columns_list,axis=1)
gold_train = gold_train.drop(columns=delate_columns_list,axis=1)
```

```python
print('Number of columns in the Training sample after removal multicollinear features: ',gold_train.shape)
print('Number of columns in Test set after removal of multicollinear features: ',gold_test.shape)
```

### Create and train the model

#### sMAPE

*The quality metric required to assess the effectiveness of the gold ore refining process is sMAPE (Symmetric Average Absolute Percentage Deviation).
Metric calculation formula:*


$$sMAPE = \frac{1}{N} \sum_{i=1}^{N}\frac{\left | y_{i}-\
\hat{y}_{i}\right |}{\left ( \left | y_{i} \right | +\left | \hat{y}_{i} \right |\right )/2} * 100\%$$


- *The value of the target feature for the object with serial number i in the sample on which the quality is measured:* $$y_{i}$$
- *Prediction value for the object with index number i, for example, in the test sample:* $$\hat{y}_{i}$$
- *Number of objects in the sample:* $$N$$
- *Summation over all sample objects (i varies from 1 to N).* $$\sum_{i=1}^{N}$$


**Formula of final sMAPE**
$$Final sMAPE = 25\%*sMAPE(rougher)+75\%*sMAPE(final)$$

```python
# Create a function that returns sMAPE:
def smape(target, prediction):
    diff = abs(target - prediction)
    add = (abs(target) + abs(prediction))/2
    smape_stage = 1/len(target)*sum(diff/add)*100    
    return smape_stage

# Create a function that returns the final sMAPE:
def final_smape(smape_r,smape_f):
    final_smape = 0.25*smape_r + 0.75*smape_f
    return final_smape
```

```python
# In each of the samples (training and test), select the features and the target feature
features_train = train_final.drop(['rougher.output.recovery','final.output.recovery'], axis=1)
target_rougher_train = train_final['rougher.output.recovery']
target_final_train = train_final['final.output.recovery']


    
features_test = test_final.drop(['rougher.output.recovery','final.output.recovery'], axis=1)
target_rougher_test = test_final['rougher.output.recovery']
target_final_test = test_final['final.output.recovery']
```

```python
# Find optimal hyperparameters for decision tree and random forest models:
linear_model = LinearRegression()

tree_params = {'max_depth': range(1,20,1)}
tree_model = DecisionTreeRegressor(random_state=12345)

forest_params = {'max_depth': range(5,20,1),'n_estimators': range(135,139,1)}
forest_model = RandomForestRegressor(random_state=12345)
```

```python
# Create an estimate that will display SMAP:
score_smape = make_scorer(smape, greater_is_better=False)
```

**Linear Regression:**

```python
%%time
cv_sc_lin_rougher = cross_val_score(LinearRegression(),
                                          features_train,
                                          target_rougher_train,
                                          cv=5, scoring=score_smape).mean()
cv_sc_lin_final = cross_val_score(LinearRegression(),
                                        features_train,
                                        target_final_train,
                                        cv=5, scoring=score_smape).mean()
smape_lin = final_smape(cv_sc_lin_rougher,cv_sc_lin_final)

print(f"Final SMAPE: {smape_lin:.3f}%")
```

**Decision Tree Regressor**

```python
%%time

tree_grid_rougher = GridSearchCV(estimator=tree_model, param_grid=tree_params, cv=5, scoring=score_smape)
tree_grid_rougher.fit(features_train, target_rougher_train)
tree_grid_rougher_best = tree_grid_rougher.best_score_

tree_grid_final = GridSearchCV(estimator=tree_model, param_grid=tree_params, cv=5, scoring=score_smape)
tree_grid_final.fit(features_train, target_final_train)
tree_grid_final_best = tree_grid_final.best_score_

smape_tree = final_smape(tree_grid_rougher_best,tree_grid_final_best)
print(f"Final SMAPE: {smape_tree:.3f}%")
```

**Random Forest Regressor**

```python
%%time

forest_grid_rougher = RandomizedSearchCV(estimator=forest_model, param_distributions=forest_params, cv=5, scoring=score_smape)
forest_grid_rougher.fit(features_train, target_rougher_train)
forest_grid_rougher_best = forest_grid_rougher.best_score_

forest_grid_final = RandomizedSearchCV(estimator=forest_model, param_distributions=forest_params, cv=5, scoring=score_smape)
forest_grid_final.fit(features_train, target_final_train)
forest_grid_final_best = forest_grid_final.best_score_

smape_forest = final_smape(forest_grid_rougher_best,forest_grid_final_best)
print(f"Final SMAPE: {smape_forest:.3f}%")
```

**Conclusion**

The best result The final SMAPE shows the **Random Forest Regressor** model fitted using the GridSearchCV hyperparameter fitting optimization method. **Total SMPE: -6.903%**

This classifier serves as a simple baseline for comparison with other more complex classifiers.
The specific behavior of the baseline is selected using the `strategy` parameter.*
- *`mean`, always predicts the mean of the training set*

```python
%%time
dummy_uniform = DummyRegressor(strategy='mean').fit(features_train, target_rougher_train)
dummy_rog = dummy_uniform.predict(features_test)
smape_dummy_roug = smape(target_rougher_test,dummy_rog)

dummy_uniform_1 = DummyRegressor(strategy='mean').fit(features_train, target_final_train)
dummy_fin = dummy_uniform_1.predict(features_test)
smape_dummy_fin = smape(target_final_test,dummy_fin)

smape_dummy = final_smape(smape_dummy_roug,smape_dummy_fin)

print(f"Total SMPE: {smape_dummy:.3f}%")
```

*Result of the adequacy test*
- Final SMPE of the true Random Forest model: 7.581%
- Final SMPE of random model: 7.779%
